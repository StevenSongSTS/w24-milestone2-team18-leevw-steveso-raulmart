```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from db_helper_functions import get_stock_news_from_db
from sklearn.decomposition import NMF, LatentDirichletAllocation, MiniBatchNMF
import matplotlib.pyplot as plt
```

```python
df = get_stock_news_from_db("AAPL")
```

```python
def plot_top_words(model, feature_names, n_top_words, title):
    fig, axes = plt.subplots(2, 5, figsize=(30, 15), sharex=True)
    axes = axes.flatten()
    for topic_idx, topic in enumerate(model.components_):
        top_features_ind = topic.argsort()[-n_top_words:]
        top_features = feature_names[top_features_ind]
        weights = topic[top_features_ind]

        ax = axes[topic_idx]
        ax.barh(top_features, weights, height=0.7)
        ax.set_title(f"Topic {topic_idx +1}", fontdict={"fontsize": 30})
        ax.tick_params(axis="both", which="major", labelsize=20)
        for i in "top right left".split():
            ax.spines[i].set_visible(False)
        fig.suptitle(title, fontsize=40)

    plt.subplots_adjust(top=0.90, bottom=0.05, wspace=0.90, hspace=0.3)
    plt.show()
```

```python
df = df[~df.article.isnull()]
tfidf = TfidfVectorizer(stop_words="english", max_df=0.95, min_df=10)
tfidf = tfidf.fit(df.article)
tfidf_mat = tfidf.transform(df.article)

tf = CountVectorizer(stop_words="english", max_df=0.95, min_df=10)
tf = tfidf.fit(df.article)
tf_mat = tfidf.transform(df.article)
```

```python
lda = LatentDirichletAllocation(
    n_components=10,
    max_iter=5,
    random_state=0,
)

lda.fit(tf_mat)
```

```python
tf_feature_names = tf.get_feature_names_out()
plot_top_words(lda, tf_feature_names, 10, "Topics in LDA model")
```
